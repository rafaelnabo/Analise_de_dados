#### Qual o Total de Vendas Por Data do Pedido?

```python
# Total de vendas para cada data de pedido
df_vendas_pedido = df.groupby('Data_Pedido')['Valor_Venda'].sum()
```

```python
# Plot utilizando Data como variavel categorica (não é série temporal)
plt.figure(figsize = (20, 6))
df_vendas_pedido.plot(x = 'Data_Pedido', y = 'Valor_Venda', color = 'green')
plt.title('Total de Vendas Por Data do Pedido')
plt.show()
```

```python
df_vendas_trimestre = df
```

```python
df_vendas_trimestre.set_index('Data_Pedido', inplace=True)
df_vendas_trimestre = df_vendas_trimestre['Valor_Venda'].resample('Q').sum()
years = df_vendas_trimestre.index.year
unique_years = years.unique()
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_years)))
year_color_map = dict(zip(unique_years, colors))
```

```python
# Plot Por trimestre
colors = bar_colors = [year_color_map[year] for year in years]
max_year = df_vendas_trimestre.groupby(years).max()

plt.figure(figsize = (20, 6))
bars = plt.bar(df_vendas_trimestre.index, df_vendas_trimestre.values, width = 30, color=colors)

for year, max_value in max_year.items():
    max_index = df_vendas_trimestre[df_vendas_trimestre == max_value].index[0]
    plt.text(max_index, max_value, f'{max_value:.2f}', ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.tight_layout()

plt.title('Total de Vendas Por Trimestre')
plt.show()
```


#### Qual Segmento Teve o Maior Total de Vendas?

```python
# Segmentos e calculamos o total de vendas
df_segmento = df.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda',
                                                                                      ascending = False)
```

```python
df_segmento.head()
```

```python
# Função para converter os dados em valor absoluto
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df_segmento['Valor_Venda'], 
        labels = df_segmento['Segmento'],
        autopct = autopct_format(df_segmento['Valor_Venda']),
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + 'R$ ' + str(round(int(sum(df_segmento['Valor_Venda']))/1000000,2)) + 'M', xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

### Quantas Vendas Receberiam 15% de Desconto?

```python
# Cria uma nova coluna de acordo com a regra definida acima
df['Desconto'] = np.where(df['Valor_Venda'] > 1000, 0.15, 0.10)
```

```python
df.head()
```

```python
# Total por cada valor da variável
df['Desconto'].value_counts()
```

```python
n_desconto = df[df['Desconto'] == 0.15].count()[0]
print(f'No Total {n_desconto} Vendas Receberiam Desconto de 15%.')
```


### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Calculando o valor de venda menos o desconto
df['Valor_Venda_Desconto'] = df['Valor_Venda'] * (1 - df['Desconto'])
```

```python
round(df.Valor_Venda.mean(), 2)
```

```python
# Filtrando as vendas antes do desconto de 15%
df_before_desc = df.loc[df['Desconto'] == 0.15, 'Valor_Venda']
```

```python
# Filtrando as vendas depois do desconto de 15%
df_after_desc = df.loc[df['Desconto'] == 0.15, 'Valor_Venda_Desconto']
```

```python
# Calcula a média das vendas antes do desconto de 15%
df_before_desc.mean()
```

```python
# Calcula a média das vendas depois do desconto de 15%
df_after_desc.mean()
```

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
#Agrupar por ano, mes e segmentos e calcular as estatisticas.
df_est_mes = df.groupby([df.index.year, df.index.month, 'Segmento'])['Valor_Venda'].agg([np.sum,np.mean,np.median])
```

```python
# Vamos extrair os níveis
anos = df_est_mes.index.get_level_values(0)
meses = df_est_mes.index.get_level_values(1)
segmentos = df_est_mes.index.get_level_values(2)
```

```python
# Plot por ano
plt.figure(figsize = (12, 6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   data = df_est_mes, 
                   y = 'mean', 
                   x = meses,
                   hue = segmentos, 
                   col = anos,
                   col_wrap = 4)
plt.show()
```

```python

# Plot por trimestre
plt.figure(figsize=(10, 6))
#Segmentação
for segmento in df.Segmento.unique():
    df_segmento = df[df.Segmento == segmento ]['Valor_Venda'].resample('Q').mean()
    plt.plot(df_segmento.index, df_segmento.values, label=segmento)


#Legendas e plot
plt.xlabel('Data')
plt.ylabel('Valor de Venda')
plt.title('Série Temporal Trimestral por Segmento')
plt.legend(title='Segmento')
plt.grid(True)
plt.show()
```

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
# Agrupar por categoria e subcategoria e calculamos a soma somente para variáveus numéricas
df_top12 = df.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
# Converter a coluna Valor_Venda em número inteiro e classificamos por categoria
df_top12 = df_top12[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
# Dataframe com categorias e subcategorias
df_top12
```

```python
# Crir outro dataframe somente com os totais por categoria
df_top12_cat = df_top12.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
# Dataframe com categorias 
df_top12_cat
```

```python
# Listas de cores para categorias
cores_categorias = ['#6699ff',
                    '#6633ff',
                    '#FF6699']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#66ccff',
                       '#66ccff',
                       '#66ccff',
                       '#66ccff',
                       '#6666ff',
                       '#6666ff',
                       '#6666ff',
                       '#6666ff',
                       '#FF99CC',
                       '#FF99CC',
                       '#FF99CC',
                       '#FF99CC']
```

```python
label_w_value = [f'{label}: R$ {round(value/1000,1)} mil' for label, value in zip(df_top12['SubCategoria'], df_top12['Valor_Venda'])]
```

```python
# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(df_top12_cat['Valor_Venda'], 
            radius = 1,
            labels = df_top12_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(df_top12['Valor_Venda'],
            radius = 0.9,
            labels = label_w_value,
            colors = cores_subcategorias, 
            labeldistance = 0.4,
            wedgeprops = dict(edgecolor = 'white'), 
            
            rotatelabels = True
            )

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.3, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(round(sum(df_top12['Valor_Venda'])/1000000, 2))+' M', xy = (-0.25, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
